```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(plt.style.available)
plt.style.use("petroff10")
data = pd.read_csv("Car_sales.csv")
data
```

```python
data.info()
```

```python
len(data[data.isnull().any(axis=1)].index)
```

```python
data["__year_resale_value"] = data["__year_resale_value"].fillna(0)
```

```python
data
```

```python
data.loc[[0,1,2,3],"Price_in_thousands"].mean()
```

```python
data.loc[[0,1,2,3],"Price_in_thousands"]=data.loc[[0,1,2,3],"Price_in_thousands"].fillna(data.loc[[0,1,2,3],"Price_in_thousands"].mean())
```

```python
data
```

```python
data.loc[[29,30,31,32,33,34,35],"Price_in_thousands"]=data.loc[[29,30,31,32,33,34,35],"Price_in_thousands"].fillna(data.loc[[29,30,31,32,33,34,35],"Price_in_thousands"].mean())
```

```python
data = data.drop(data.index[35])
```

```python
data = data.dropna()
```

```python
data
```

```python
sbm = data.groupby("Manufacturer")["Sales_in_thousands"].sum()
```

```python
sbm.plot.bar()
```

```python
sns.barplot(data=sbm,orient="h",gap=1)
```

```python
sbm.plot.pie(title="Share in market sales",figsize=(7,7))
```

```python
price_dep = data.groupby("Manufacturer")[["Price_in_thousands","__year_resale_value"]].sum()
```

```python
price_dep.plot.bar()
```

```python
value_fac = data.groupby("Manufacturer")["Power_perf_factor"].sum().sort_values()
```

```python
value_fac.plot.bar()
```

```python
price_perf = data.groupby("Manufacturer")[["Price_in_thousands","Power_perf_factor"]].sum()
```

```python
price_perf.plot.bar()
```

```python
vtype = data.groupby("Vehicle_type")["Manufacturer"].count()
vtype.plot.pie(title="Vehicle sales by type",figsize=(7,7),autopct="%1.1f%%")
```

```python
hp_fuel = pd.DataFrame({
    "Manufacturers":data["Manufacturer"]
})
hp_fuel["factor"] = data["Horsepower"] / data["Fuel_efficiency"]
hp_fuel.groupby("Manufacturers")["factor"].mean().plot.bar(title="Horsepower to fuel efficiency quotient")
```
